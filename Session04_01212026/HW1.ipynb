```python
#create a list of squares of even numbers from 1 to 20
def squares_of_even_numbers():
        return [x**2 for x in range(1, 21) if x % 2 == 0]

if __name__ == "__main__":
    print(squares_of_even_numbers()) 
```

```python
def student_score_dictionary():
    return {
        "Alice": 85,
        "Bob": 92,
        "Charlie": 78,
        "David": 90,
        "Eva": 88
    }

if __name__ == "__main__":     
    scores = student_score_dictionary()
    for student, score in scores.items():
        if score >= 80:
            print(f"{student}: {score}")
```

```python
def test_squares_of_even_numbers():
    expected = [4, 16, 36, 64, 100, 144, 196, 256, 324, 400]
    actual = squares_of_even_numbers()
    assert actual == expected, f"Fail: Expected {expected}, but got {actual}"
    print("Pass")

test_squares_of_even_numbers()
```
