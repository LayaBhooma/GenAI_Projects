```python
# 3. Write a function that counts how many words occur more than once in a given paragraph.

def count_repeated_words(paragraph):
    from collections import Counter
    words = paragraph.split()
    word_counts = Counter(words)
    repeated_words = {word: count for word, count in word_counts.items() if count > 1}
    return repeated_words
paragraph = "This is a test. This test is only a test."
print(count_repeated_words(paragraph))
```

```python
# 4. Write a function that takes a list of integers and returns a new list with only the prime numbers.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def get_primes(numbers):
    return [n for n in numbers if is_prime(n)]
print(get_primes([10, 15, 23, 42, 7, 3, 4, 5]))

```

```python
# 6. Using ‘SentenceTransformer’, encode two sentences and compute cosine similarity.
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
sentence1 = "This is a sample sentence."
sentence2 = "This is another example sentence."
embedding1 = model.encode(sentence1, convert_to_tensor=True)
embedding2 = model.encode(sentence2, convert_to_tensor=True)
cosine_similarity = util.pytorch_cos_sim(embedding1, embedding2)
print(f"Cosine Similarity: {cosine_similarity.item()}")
```

```python
# 7. (Bonus) Write a function that creates embeddings for each sentence in a text file and saves them as a CSV.
import csv
def create_embeddings_csv(input_file, output_file):
    from sentence_transformers import SentenceTransformer
    model = SentenceTransformer('all-MiniLM-L6-v2')
    
    with open(input_file, 'r', encoding="utf-8") as f:
        sentences = f.readlines()
    
    embeddings = model.encode(sentences)
    
    with open(output_file, 'w', newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(['Sentence'] + [f'Embedding_{i}' for i in range(len(embeddings[0]))])
        for sentence, embedding in zip(sentences, embeddings):
            writer.writerow([sentence.strip()] + embedding.tolist())
    
    print(f"Embeddings saved to {output_file}")

create_embeddings_csv('./What_SpeakEQ_Can_Learn.txt', 'embeddings.csv')
```

```python
#Create an embedding representation for each document.

#Embedding values may be hardcoded 
from sentence_transformers import SentenceTransformer

def create_hardcoded_embeddings(docs):
    # These represent the 'coordinates' of your text in vector space.
    # In a real scenario, these would be 384 or 768 dimensions long.
    hardcoded_map = {
        1: [0.12, -0.05, 0.88, -0.21], # Representing 'Company policies'
        2: [0.09, 0.11, -0.45, 0.72],  # Representing 'Code of conduct'
        3: [-0.67, 0.34, 0.12, -0.08]  # Representing 'Student rules'
    }
    return [hardcoded_map[doc['id']] for doc in docs]


def create_embeddings(documents):
    return create_hardcoded_embeddings(documents)

documents = [
    {"id":1, "text":"Company policies are designed to ensure a safe and productive work environment. All staff members must receive adequate training on these safety standards to protect themselves and others. Management is ultimately responsible for maintaining equipment and infrastructure in good repair."},
    {"id":2, "text":"Employees must adhere to the code of conduct at all times. This includes maintaining professional behavior and avoiding any conflicts of interest that could harm the company's reputation. Failure to comply with these ethical standards may result in formal disciplinary action."},
    {"id":3, "text":"Student rules include attending classes regularly and submitting assignments on time. Persistent unexcused absences can lead to a significant reduction in a student's final grade. It is essential to follow all formatting and submission guidelines to ensure work is accepted for evaluation."}
]

embeddings = create_embeddings(documents)

# Store embeddings in a Python structure (e.g., dictionary or list).
# Store results in a dictionary for easy lookup by ID
document_embeddings = {
    doc['id']: emb for doc, emb in zip(documents, embeddings)
}

print(f"Embedding for Doc 1: {document_embeddings[1]}")
```
