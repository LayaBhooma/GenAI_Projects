```python
#Event Extraction from Image (Flyer or Poster)
import os
from dotenv import load_dotenv

import base64
load_dotenv(override=True, dotenv_path="../.env.local")

# Initialize the client
from openai import OpenAI
my_api_key = os.getenv("OPENAI_API_KEY")

client = OpenAI(api_key=my_api_key)

def encode_image(image_path):
    """Encodes a local image to base64 format."""
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# 1. Path to the local image
image_path = "./data/OwlsEvent.JPG" 

# 2. Convert to base64
base64_image = encode_image(image_path)
```

```python
# 3. Request data extraction
response = client.chat.completions.create(
    model="gpt-4o",
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Extract Event name, Date and time, Location, Organizer, and Event summary(2–3 lines) and format it as JSON."},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base64_image}"
                    }
                }
            ]
        }
    ],
    max_tokens=500
)
```
