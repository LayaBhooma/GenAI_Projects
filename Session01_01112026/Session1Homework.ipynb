```python
#%pip install python-dotenv
import os
from dotenv import load_dotenv
load_dotenv(override=True, dotenv_path="../.env.local")

#reverse a String
def reverse_string_loop(s):
    """Reverses a string using a for loop."""
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str

#Check for Palindrome
def is_palindrome(s):
    cleaned = ''.join(c.lower() for c in s if c.isalnum())
    return cleaned == cleaned[::-1]

# Find a missing number in a list
def find_missing_number(nums, n):
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)
    return expected_sum - actual_sum

#Count Word Frequency
 #Given a sentence, return a dictionary with word counts.
 #Example: "AI makes AI easier" â†’ { "AI": 2, "makes": 1, "easier": 1 }.
def count_word_frequency(sentence):
    words = sentence.split()
    frequency = {}
    for word in words:
        word = word.lower()
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency

#FizzBuzz
# Print numbers from 1 to 50. For multiples of 3, print "Fizz", for multiples of 5 print "Buzz", and for multiples of both, print "FizzBuzz".
def fizz_buzz():
    for i in range(1, 51):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)
#process
original = input("Enter a string")
reversed_str = reverse_string_loop(original)
print(f"Original: {original}, Reversed: {reversed_str}")
print(f"Is Palindrome: {is_palindrome(original)}")

nums_len= int(input("Enter the range of numbers (n): "))
user_input = input(f"Enter {nums_len-1} numbers separated by space: ")  
nums = [int(x) for x in user_input.split()]  

missing_number = find_missing_number(nums, nums_len)
print(f"Missing number in the list is: {missing_number}") 
sentence = input("Enter a sentence: ")
word_freq = count_word_frequency(sentence)
print("Word Frequency:", word_freq)  
fizz_buzz()
```
