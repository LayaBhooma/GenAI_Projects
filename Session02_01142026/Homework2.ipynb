```python
# Write a Python script that Creates a list of dictionaries, where each dictionary represents a movie:

listOfMovies = []
while True:
    movie = input("Enter movie details in the format Title,Genre,Year: , exit to stop: ")
    if movie.lower() == "exit":
        break
    listOfMovies.append(movie)

#collection = [
#    {"title": "Inception", "genre": "Sci-Fi", "year": 2010},
#    {"title": "The Matrix", "genre": "Action", "year": 1999},
#    {"title": "Coco", "genre": "Animation", "year": 2017},
#]
```

```python
#Prints each item in this format : "Inception" is a Sci-Fi movie released in 2010.
print("\nMovie List:")
for movie in listOfMovies:
    title, genre, year = movie.split(",")
    print(f'"{title}" is a {genre} movie released in {year}.')  
```

```python
#Adds a search feature:  Ask the user for a genre and print all movies that match the entered genre.
search_genre = input("\nEnter a genre to search for movies: ")
print(f"\nMovies in the genre '{search_genre}':")
for movie in listOfMovies:
    title, genre, year = movie.split(",")
    print(f"DEBUG: D{genre.lower()}D, D{search_genre.lower()}D)")
    if (genre.lower() == search_genre.lower()):
        print(f'"{title}" is a {genre} movie released in {year}.') 
    
```

```python
scores = [78, 85, 62, 90, 56, 99, 74]

#Calculate and print the average score.
average_score = sum(scores) / len(scores)
print(f"Average score: {average_score}")    

#Print all scores greater than 80.
print(f"Scores greater than 80:")
for score in scores:
    if score > 80:
        print(f"  {score}")

#Replace scores below 60 with the word "Fail".
for i in range(len(scores)):
    if scores[i] < 60:
        scores[i] = "Fail"
print("Updated scores list:", scores)
```
